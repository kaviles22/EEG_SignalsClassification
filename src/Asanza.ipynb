```python
df= pd.read_csv('../DatosBrutos/Frecuencia/PB1.csv')
```

```python
senales = df[df.columns[9:11]]
etiqueta = df[df.columns[21:22]]
datos = pd.concat([senales, etiqueta], axis=1)
datos
```

```python
c = 6
for e in range(len(datos.index)):
    if str(datos['MarkerValueInt'][e]) == 'nan':
        datos['MarkerValueInt'][e] = c
    else:
        c = datos['MarkerValueInt'][e]
```

```python
datos_nuevos = datos[datos['MarkerValueInt'] != 6]
datos_nuevos
```

```python
rojo = datos_nuevos[datos_nuevos['MarkerValueInt'] == 1]
azul = datos_nuevos[datos_nuevos['MarkerValueInt'] == 2]
verde = datos_nuevos[datos_nuevos['MarkerValueInt'] == 3]
morado = datos_nuevos[datos_nuevos['MarkerValueInt'] == 4]
baseline = datos_nuevos[datos_nuevos['MarkerValueInt'] == 5]
```

```python
datos_nuevos.to_csv('../DatosProcesados/datos.csv')
df = pd.read_csv('../DatosProcesados/datos.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis =1)
df
```

```python
etiquetas = ['Rojo', 'Azul', 'Verde', 'Morado', 'Baseline']
contadores = [0, 0, 0, 0, 0]
c = 5
inicio = 0
num_archivos = 0
for e in range(len(df.index)):
    if int(df['MarkerValueInt'][e]) != c:
        archivo = df.iloc[inicio:e-1, :]
        print(archivo)
        archivo.to_csv(f'../DatosProcesados/{etiquetas[c - 1]}/ejemplo{contadores[c -1]}.csv', index = False)
        contadores[c-1] += 1
        inicio = e
        c = int(df['MarkerValueInt'][e]) 
        num_archivos +=1
print(num_archivos)
!rm ../DatosProcesados/datos.csv
```
